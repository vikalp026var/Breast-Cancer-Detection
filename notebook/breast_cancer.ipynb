```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report
from sklearn.model_selection import GridSearchCV
```

```python
df
```

```python
Y=df['diagnosis']
X=df.drop(columns=['diagnosis'],axis=1)
```

```python
X.shape
```

```python
Y.shape
```

```python
scaler=StandardScaler()
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
X_scaled
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_scaled,Y,test_size=0.33,random_state=42)
```

```python
X_train.shape
```

```python
X_train
```

```python
models={
    'LogisticRegression':LogisticRegression(),
    'GaussianNB':GaussianNB(),
    'DecisionTreeClassifier':DecisionTreeClassifier()
    
}
```

```python
for model_name, model in models.items():
    model.fit(X_train, y_train)  # Train the model
    y_pred = model.predict(X_test)  # Make predictions
    accuracy = accuracy_score(y_test, y_pred)  # Calculate accuracy
    print(f"Accuracy of {model_name} is {accuracy:.2f}")
```

```python
model=LogisticRegression(solver = 'liblinear', penalty = 'l2')
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
para_grid={
    'C':np.linspace(0.001, 1, 20), 'fit_intercept':[True, False]
}
```

```python
clf=GridSearchCV(model,param_grid=para_grid,cv=10,verbose=3)
```

```python
clf
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred_tun=clf.predict(X_test)
```

```python
accuracy_score(y_pred_tun,y_test)
```

```python
classification_report(y_test,y_pred_tun)
```
